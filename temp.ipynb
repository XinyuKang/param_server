```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
total_gradient_time = []
total_time = []
total_acc=[]
std_acc=[]
for idx in [1,2,4,6,8,10]:
    s10w1_1 = json.load(open("/home/bwang/classes/CS262/param_server/results/servers_1_workerss_"+str(idx)+".json"))
    tot_time = (np.array(s10w1_1["total_time"][1:]))
    tot_grad_time = (np.array(s10w1_1["gradient_time"][1:]))
    tot_acc = (np.array(s10w1_1["accuracy"][0:]))
    
    plt.plot(np.arange(0,400,5), tot_acc, label=str(idx)+" workers")
    total_gradient_time.append(np.mean(tot_grad_time))
    total_time.append(np.mean(tot_time))

    val = np.max(s10w1_1["accuracy"])
    total_acc.append(np.mean([val]))
    std_acc.append(np.std(val))

plt.xlabel("Number of iterations")
plt.ylabel("Final accuracy")
plt.title("Accuracy vs. number of iterations with 1 server")
plt.legend()
```

```python
total_gradient_time_w = []
total_time_w=[]
total_acc_w =[]
std_acc_w =[]
for idx in [1,2,4,6,8,10]:
    s10w1_1 = json.load(open("/home/bwang/classes/CS262/param_server/results/servers_1_workerss_"+str(idx)+".json"))
    tot_time = (np.array(s10w1_1["total_time"][1:]))
    total_acc = (np.array(s10w1_1["accuracy"]))
    tot_grad_time = (np.array(s10w1_1["gradient_time"][1:]))

    plt.plot(tot_grad_time, label=str(idx)+" workers")
    total_gradient_time_w.append(np.mean(tot_grad_time))
    total_time_w.append(np.mean(tot_time))
    
    val = [np.max(s10w1_1["accuracy"])]
    total_acc_w.append(np.mean([val]))
    std_acc_w.append(np.std(val))
    
plt.legend()
plt.ylabel("Time")
plt.xlabel("Iteration")
plt.title("Time vs. number of iterations with 1 server")
```
